```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
#load the data from the csv file into dataframe
df=pd.read_csv('/content/data (3).csv')
one_hot_encoded=pd.get_dummies(df['category'])
label_encoder=LabelEncoder()
label_encoded=label_encoder.fit_transform(df['category'])
print("original data:")
print(df)
print("\nOne-hot encoded data:")
print(one_hot_encoded)
print("\nLabel encoded data:")
print(label_encoded)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler=StandardScaler()
standardized_data=scaler.fit_transform(df[['feature1']])
min_max_scaler=MinMaxScaler()
min_max_scaled_data=min_max_scaler.fit_transform(df[['feature1']])
print("original data:")
print(df)
print("\nStandardized data:")
print(standardized_data)
print("\nMin-Max scaled data:")
print(min_max_scaled_data)
```
